```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("darkgrid")
import warnings 
warnings.filterwarnings("ignore")

# Machine Learning Libraries 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
df = pd.read_csv(r"H:\Programming\Dataset\Machine Learning\placement_regression.csv")
df.head()
```

```python
sns.pairplot(df, height=2, aspect=2.5) 
plt.show()
```

```python
x = df[['cgpa']]
y = df['package']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# kernel : {'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'}
```

```python
svr = SVR(kernel='linear', degree=3)
svr.fit(x,y)
```

```python
train_score = svr.score(x_train, y_train)
test_score = svr.score(x_test, y_test)

result = pd.DataFrame([[train_score, test_score]], columns=['Train_Score', 'Test_Score'])
result
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='cgpa', y='package', data=df)
sns.scatterplot(x='cgpa', y=svr.predict(x), data=df, color='red')
# plt.savefig("line.png")
plt.show()
```
