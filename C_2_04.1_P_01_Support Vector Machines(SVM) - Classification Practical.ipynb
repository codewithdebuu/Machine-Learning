```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("darkgrid")
import warnings 
warnings.filterwarnings("ignore")

# Machine Learning Libraries 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_excel(r"H:\Programming\Dataset\Machine Learning\Placement (2).xlsx")
df.head()
```

```python
df.info()
```

```python
sns.pairplot(df, hue="Placement", height=2, aspect=2.5) 
plt.show()
```

```python
x = df.drop(columns="Placement", axis=0)
y = df['Placement']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# kernel={'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'}
```

```python
svc = SVC(kernel='rbf', random_state=42)
svc.fit(x,y)
```

```python
train_score = svc.score(x_train, y_train)
test_score = svc.score(x_test, y_test)

result = pd.DataFrame([[train_score, test_score]], columns=['Train_Score', 'Test_Score'])
result
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=svc)
# plt.savefig('svc.png')
plt.show()
```
