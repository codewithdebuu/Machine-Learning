```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.set_style("darkgrid")

import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"H:\Programming\Files\Machine Learning\Files\Loan Dataset.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
# Fill NaN Values 
for i in df.columns:
    df[i].fillna(df[i].mode()[0], inplace=True)

df.isnull().sum()
```

```python
df.describe()
```

```python
sns.distplot(df['CoapplicantIncome'])
plt.show()
```

```python
sns.histplot(df['CoapplicantIncome'], kde=True)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

mms.fit(df[['CoapplicantIncome']])
```

```python
df['CoapplicantIncome_mms'] = mms.transform(df[['CoapplicantIncome']])
df.head(3)
```

```python
plt.figure(figsize=(15,5)) 
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df['CoapplicantIncome'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(df['CoapplicantIncome_mms'])

plt.show()
```
