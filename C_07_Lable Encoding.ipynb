# All in one

```python
# Import pandas for data manipulation
import pandas as pd 

# Import LabelEncoder for encoding categorical labels
from sklearn.preprocessing import LabelEncoder

# Select only the 'Property_Area' column from the DataFrame
df = df[['Property_Area']]

# Filling NaN Values by column mode 
df['Property_Area'].fillna(df['Property_Area'].mode()[0], inplace=True)

# Initialize the LabelEncoder
le = LabelEncoder()

# Fit the encoder on the 'Property_Area' column
le.fit(df['Property_Area'])

# Transform the 'Property_Area' column and store it in a new 'label' column
df['label'] = le.transform(df['Property_Area'])

# Display the first 5 rows of the DataFrame
df.head()
```

### Shortcut

```python
df = df[['Property_Area']]

df['Property_Area'].fillna(df['Property_Area'].mode()[0], inplace=True)

le = LabelEncoder()
df['label'] = le.fit_transform(df['Property_Area'])

df.head()
```
