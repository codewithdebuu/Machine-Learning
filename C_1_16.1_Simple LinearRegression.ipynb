```python
# Import Important Python Libraries
import pandas as pd 

# Visualization Libraries for EDA
import seaborn as sns 
import matplotlib.pyplot as plt 

# Machine Learning Libraries 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
df = pd.read_csv(r"H:\Programming\Dataset\Machine Learning\placement.csv")
df.head()
```

```python
#scatterplot
sns.scatterplot(df)
plt.show()
```

```python
sns.scatterplot(x="cgpa", y="package",data=df)

plt.show()
```

```python
# Value assigning
x = df[['cgpa']]
y = df['package']

x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
# LINEAR REGRESSION MODEL
lr.fit(x_train, y_train)
```

```python
print(f"Accuracy: {lr.score(x_test, y_test)}")
```

```python
y_prd = lr.predict(x)
plt.figure(figsize=(6,4))

# Plot original data as scatter plot
sns.scatterplot(x='cgpa', y='package', data=df)

# Plot regression line
plt.plot(df['cgpa'], y_prd, color='red')  

plt.legend(["Original_data", "Predict_line"])
plt.show()
```
