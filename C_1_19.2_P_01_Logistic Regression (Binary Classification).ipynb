# Train Test Split

```python
x = df[['Age']]
y = df['Purchased']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
print("Successfully completed.")
```

# LogisticRegression

```python
lor = LogisticRegression()
lor.fit(x_train, y_train)
```

```python
print(f"Model Accuracy: {lor.score(x_test, y_test)*100: .2f}%")
```

```python
lor.predict([[45]])
```

```python
sns.scatterplot(x='Age', y='Purchased', data=df, hue='Purchased')
sns.lineplot(x='Age', y=lor.predict(x), data=df, color='green')
plt.show()
```
