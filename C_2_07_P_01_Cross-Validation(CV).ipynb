```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import LeaveOneOut, LeavePOut, KFold, StratifiedKFold, cross_val_score
```

```python
df = pd.read_csv(r"H:\Programming\Dataset\Machine Learning\placement_regression.csv")
df.head()
```

```python
df.info()
```

```python
new_df = df.iloc[:10,:]
new_df.info()
```

```python
new_df.head(3)
```

```python
x=new_df[['cgpa']]
y=new_df['package']
```

# LinearRegression Model Cross Validate (cv)

```python
p = cross_val_score(LinearRegression(), x, y, cv=5)
p.sort()  # min to max
p
```

# Leave-One-Out

```python
lo = LeaveOneOut()

for train, test in lo.split(x,y): 
    print(train, test)
```

# Leave-P-Out

```python
lp = LeavePOut(p=2)

for train, test in lp.split(x,y): 
    print(train, test)
```

```python
p2 = cross_val_score(LinearRegression(), x, y, cv=lp)
p2.sort()  # min to max
p2
```

# KFold 

```python
kf = KFold(n_splits=3)

for train, test in kf.split(x,y): 
    print(train, test)
```

```python
p3 = cross_val_score(LinearRegression(), x, y, cv=kf)
p3.sort()  # min to max
p3
```

# Model final result (min-max)

```python
a = [p.min().item(), p.max().item()]
b = [p2.min().item(), p2.max().item()]
c = [p3.min().item(), p3.max().item()]
idx = ["Normal", "LeavePOut", "KFold"]

result = pd.DataFrame([a, b, c], columns=["Min", "Max"], index=idx)
result
```
