```python
# Import necessary libraries
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
import numpy as np

# Features and target
x = df.iloc[:, :-1]  # all columns except the last
y = df['Outcome']    # target column

# Initialize logistic regression
lr = LogisticRegression(max_iter=1000)

# Define cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize backward feature selector
sfs = SequentialFeatureSelector(
    lr,
    n_features_to_select="auto",
    direction='forward',    # Forward selectio
    #direction='backward',   # Backward selection
    scoring='accuracy',
    cv=cv
)

# Fit feature selector
sfs.fit(x, y)

# Get the selected feature names
best_features = x.columns[sfs.get_support()]

# Print selected features
print("Number of best features:", len(best_features))
print("Best features:", list(best_features))

# Evaluate cross-validated accuracy using only selected features
cv_scores = cross_val_score(lr, x[best_features], y, cv=cv, scoring='accuracy')
print("Cross-validated accuracy:", np.mean(cv_scores))
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from mlxtend.feature_selection import SequentialFeatureSelector

# Split features and target
X = df.iloc[:, :-1]  # all columns except the last
y = df['Outcome']    # target column

# Initialize Logistic Regression classifier
lr = LogisticRegression(max_iter=1000)

# ---------------- Forward Feature Selection ----------------
sfs_forward = SequentialFeatureSelector(
    lr,
    k_features='best',
    forward=True,
    floating=False,
    scoring='accuracy',
    cv=5,       # 5-fold cross-validation without StratifiedKFold
    n_jobs=-1
)
sfs_forward.fit(X, y)

best_features_forward = X.columns[list(sfs_forward.k_feature_idx_)]

print("Forward Selection:")
print("Best number of features:", len(best_features_forward))
print("Best feature names:", list(best_features_forward))
print("Best CV accuracy:", sfs_forward.k_score_)

# ---------------- Backward Feature Elimination ----------------
sfs_backward = SequentialFeatureSelector(
    lr,
    k_features='best',
    forward=False,
    floating=False,
    scoring='accuracy',
    cv=5,       # 5-fold cross-validation without StratifiedKFold
    n_jobs=-1
)
sfs_backward.fit(X, y)

best_features_backward = X.columns[list(sfs_backward.k_feature_idx_)]

print("\nBackward Elimination:")
print("Best number of features:", len(best_features_backward))
print("Best feature names:", list(best_features_backward))
print("Best CV accuracy:", sfs_backward.k_score_)
```
